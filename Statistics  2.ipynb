```python
#distance matrics 
from scipy.spatial import distance 
x = [3,6,9]
y = [1,0,1]
print(distance.euclidean(x,y))
```

```python
#city bloak 
from scipy.spatial import distance 
x = [3,6,9]
y = [1,0,1]
print(distance.cityblock(x,y))
```
